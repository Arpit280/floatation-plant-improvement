```python


# Step 1: Reduce the dataset to every 100th row
reduced_df = df.iloc[::100, :].reset_index(drop=True)

# Step 2: Calculate the number of rows per day and repeat the day values
rows_per_day = len(reduced_df) // 180  # Approximately 40 rows per day
day_values = [day for day in range(1, 181) for _ in range(rows_per_day)]
remaining_rows = len(reduced_df) - len(day_values)  # Handle any leftover rows

# Extend day_values to cover remaining rows if any
day_values.extend([180] * remaining_rows)
reduced_df['day'] = day_values

# Step 3: Calculate the 'week' column
reduced_df['week'] = ((reduced_df['day'] - 1) // 7) + 1

# Display the first few rows to verify
print(reduced_df.head(200))  # Check the first few days
```
